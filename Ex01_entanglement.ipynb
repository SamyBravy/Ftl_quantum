```python
import qiskit
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
```

![ex00_Theory_1](Theory/ex01_Theory_1.jpeg)

```python
qc = QuantumCircuit(2) # 2 qbits
qc.h(0) # Hadamard on first qbit
qc.cx(0, 1) # cnot between first and second qbit

state = Statevector.from_instruction(qc)

qc.measure_all() # State Projector Measure

print("Circuit:")
display(qc.draw(output='mpl'))

print("Bloch Sphere:")
plot_bloch_multivector(state) # Entangled
```

![ex00_Theory_2](Theory/ex01_Theory_2.jpeg)

```python
shots = 500

backend = AerSimulator()
job = backend.run(qc, shots=shots)
result = job.result()

counts = result.get_counts()
probabilities = {state : count / shots for state, count in counts.items()}
print(f"Probabilities {shots} shots:")
plot_histogram(probabilities)
```
